```python
from piva.data_loader import DataloaderBloch as dl
from piva.working_procedures import get_step
from pathlib import Path
from os.path import getmtime
from datetime import datetime
import pandas as pd
```

```python
# Specify path
path = ''
# open files and save scans' names
files=[]
fnames=[]
date_and_time = []
for fileobj in sorted(Path(path).iterdir(), key=getmtime):
    filename = str(fileobj)
    try:
        tmp_dl = dl()
        files.append(tmp_dl.load_data(filename))
        fnames.append(fileobj.name)
        date_and_time.append(datetime.fromtimestamp(getmtime(filename)))
        print(filename)
    except (ValueError,  OSError, NotImplementedError) as e:
        continue
```

```python
# collect the data that are actually interesting
temp=[]
hv=[]
grating=[]
polarization=[]
PE=[]
lens_mode=[]
anal_slit=[]
acq_mode=[]
DT=[]
x_pos=[]
y_pos=[]
z_pos=[]
theta=[]
phi=[]
tilt=[]
scan_type=[]
energy_range=[]
energy_step=[]
FE_slit=[]
V_slit=[]
comments=[]
for idx, file in enumerate(files):
    temp.append(file.temp)
    hv.append(file.hv)
    grating.append(' ')
    polarization.append(file.polarization)
    PE.append(file.PE)
    lens_mode.append(file.lens_mode)
    anal_slit.append(file.anal_slit)
    acq_mode.append(file.acq_mode)
    DT.append(file.DT)
    x_pos.append(file.x)
    y_pos.append(file.y)
    z_pos.append(file.z)
    theta.append(file.theta)
    scan_type.append(file.scan_type)
    phi.append(file.phi)
    tilt.append(file.tilt)
    energy_range.append('{:.2f} -> {:.2f}'.format(file.zscale.min(), file.zscale.max()))
    energy_step.append(get_step(file.zscale))
    FE_slit.append(file.FE)
    V_slit.append(file.exit_slit)
del files
```

```python
# print what you want
d = {
     'scan name':fnames,
     'date and time':date_and_time,
     'T, K':temp,
     'hv, eV':hv,
     'PE, eV':PE,
     'PGM':grating,
     'polarization':polarization,
     'FE, mm':FE_slit,
     'V slit, $\mu$m':V_slit,
     'x, mm':x_pos,
     'y, mm':y_pos,
     'z, mm':z_pos,
     'theta, deg':theta,
     'phi, deg':phi,
     'tilt (step), deg':tilt,
     'scan type':scan_type,
     'lens mode':lens_mode,
     'anal slit':anal_slit,
     'E range, eV':energy_range,
     'E step, meV':energy_step,
     'acq mode':acq_mode,
     'DT, ms':DT
    }
df = pd.DataFrame(data=d)#.sort_values(by='date and time')
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 500)
df
```
