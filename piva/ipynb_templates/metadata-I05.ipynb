```python
from piva.data_loader import Dataloaderi05 as dl
from piva.working_procedures import get_step
from pathlib import Path
from os.path import getmtime
from datetime import datetime
import pandas as pd
```

```python
# Specify path
path = ''
# open files and save scans' names
files=[]
fnames=[]
date_and_time = []
for fileobj in sorted(Path(path).iterdir(), key=getmtime):
    filename = str(fileobj)
    try:
        tmp_dl = dl()
        files.append(tmp_dl.load_data(filename))
        fnames.append(fileobj.name)
        date_and_time.append(datetime.fromtimestamp(getmtime(filename)))
        print(filename)
    except (ValueError,  OSError, NotImplementedError, KeyError) as e:
        continue
```

```python
# collect the data that are actually interesting
temp=[]
hv=[]
grating=[]
polarization=[]
PE=[]
lens_mode=[]
anal_slit=[]
acq_mode=[]
DT=[]
x_pos=[]
y_pos=[]
z_pos=[]
theta=[]
phi=[]
tilt=[]
scan_type=[]
energy_range=[]
energy_step=[]
FE_slit=[]
exit_slit=[]
comments=[]
for idx, file in enumerate(files):
    temp.append(file.temp)
    hv.append(file.hv)
    grating.append(' ')
    polarization.append(file.polarization)
    PE.append(file.PE)
    lens_mode.append(file.lens_mode)
    anal_slit.append(file.anal_slit)
    acq_mode.append(file.acq_mode)
    DT.append(file.DT)
    x_pos.append(file.x)
    y_pos.append(file.y)
    z_pos.append(file.z)
    theta.append(file.theta)
    scan_type.append(file.scan_type)
    phi.append(file.phi)
    tilt.append(file.tilt)
    energy_range.append('{:.2f} -> {:.2f}'.format(file.zscale.min(), file.zscale.max()))
    energy_step.append(get_step(file.zscale))
    FE_slit.append(file.FE)
    exit_slit.append(file.exit_slit)
del files
```

```python
# print what you want
d = {'scan name':fnames,
     'date and time':date_and_time,
     'T, K':temp,
     'hv, eV':hv,
     'PGM':grating,
     'polarization':polarization,
     'exit slit, mm':exit_slit,
     'x, mm':x_pos,
     'y, mm':y_pos,
     'z, mm':z_pos,
     'tilt, deg':tilt,
     'azimuth, deg':phi,
     'polar, deg':theta,
     'PE, eV':PE,
     'lens mode':lens_mode,
     'Ek range, eV':energy_range,
     'E step, meV':energy_step,
     'acq mode':acq_mode,
     'DT, s':DT
    }
df = pd.DataFrame(data=d).sort_values(by='date and time')
pd.set_option('display.max_columns', 50)
df
```
